```python
path='filestoimport'
```

```python
path=path + "/*.xlsx"
path
```

```python
list_of_dfs = []
# append df to data in each for
data = []
# for in excel file
for filepath in glob.glob(path):
    xl = pd.ExcelFile(filepath)
 
 
# for in sheet of excel file
    for sheet_name in xl.sheet_names:
        df = xl.parse(sheet_name, usecols='A,D,N,B,C,E,F,G,H,I,J,K,L,M', header=0)
        df.columns = df.columns.str.replace(' ', '')
 
        df['sheetname'] = sheet_name  # this adds `sheet_name` into the column
         
        # using basename function from os
        # module to print file name
        file_name = os.path.basename(filepath)
        df['sourcefilename'] = file_name
        data.append(df)
        list_of_dfs.append([sheet_name,file_name])
       
    # Combine all DataFrames into one
        bigframe = pd.concat(data, ignore_index=True)
```
